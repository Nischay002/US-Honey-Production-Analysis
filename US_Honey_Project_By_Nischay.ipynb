#BOTTOM 5 and TOP 5 STATES IN US FOR HONEY PRODUCING

```python
#GROUPBY
states_production = df.groupby('state')['production'].sum().reset_index()
states_production
```

```python
top_5 = states_production.sort_values(by='production',ascending = False).head()
```

```python
top_5
```

```python
sns.barplot(x='state',y='production',data=top_5,palette= 'dark')
plt.show()
```

```python
bottom_5 = states_production.sort_values(by='production',ascending = True).head()
bottom_5
```

```python
sns.barplot(x='state',y='production',data=bottom_5,palette= 'dark')
plt.show()
```

#Which was the year when production of Honey in wholeUS was the highest?

```python
highest_pro = df.groupby('year')['production'].sum().reset_index()
highest_pro
```

```python
highest_pro.sort_values(by = 'production', ascending = False).head(1)
```

#From the above inference we get the production was highest in the year 2000, now let infer which state was having highest contribution in that year

```python
highest_state = df.groupby(['state','year'])['production'].sum().reset_index()
```

```python
highest_state
```

```python
state_2000 = highest_state.loc[(highest_state['year']== 2000)]
state_2000
```

```python
state_2000.sort_values(by='production',ascending = False).head(1)
```
